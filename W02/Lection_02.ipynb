## Переменные

Переменные в Python хранят ссылку на объект в памяти и информацию о его типе. 

```python
# переменная содержит ссылку на объект типа int
# объект хранит значение 5
a = 5
print type(a)
```

```python
# проверка на соответствие типов
type(a) is int
```

Основные встроенные типы данных в Python разбирались в лекции 1:

- `int`
- `float`
- `str`
- `list`
- `tuple`
- `dict`
- `set`

Тип переменной определяется автоматически при присваивании значения и изменяется следующим присваиванием.

```python
some_input = 'Строка с текстом, имеет в данном случае тип str'
print 'some_input имеет тип:', type(some_input)
# вызываем метод split - он возвращает список строк
some_input = some_input.split()
print 'some_input имеет тип:', type(some_input)
```

При присваивании в переменной сохраняется не значение, а ссылка на объект. Это значит, что изменяя одну переменную 
вы можете изменить и другую:

```python
some_array = ['first item', 'second item', 'third item']
new_array = some_array
new_array.append('fourh item')
print 'Первый массив:', some_array
print 'Новый массив:', new_array
```

Как видно, значение добавилось в обе переменных. Это случилось из-за того, что переменные хранят ссылку на один и тот же объект в памяти. Проверить, что две переменные содержат ссылку на один один объект можно с помощью оператора is

```python
some_array is new_array
```

## Условия

Условия позволяют управлять потоком выполнения команд. 

Условный оператор `if` требует создания блока команд, а значит наличия отступов:

```python
if len(some_array) > 3:
    print 'Любой блок в Python начинается с двоеточия'
    print 'Затем со следующей строки отступы увеличиваются на 4 пробела'
elif len(some_array) < 2:
    print 'блок elif позволяет проверить новое условие, если условие в if ложно'
else:
    print 'блок else выполняется в случае, если ни одно из условий выше не было истиным'
print 'После окончания блока - отступы возвращаются к предыдующему количеству'
```

## Циклы

В Python используются два типа циклов: 

- с предусловием (`while`)
- цикл для каждого элемента последовательности (`for`)

Цикл `while` всегда имеет три элемента:

- инициализация счётчика (присвоение начального значения)
- проверка условия
- изменение счётчика

```python
i = 10
while i > 0:
    print i,
    i -= 1
```

Проверка условия может происходить не в заголовке, а в теле цикла. В этом случае нужно использовать инструкции continue и break

```python
index = 0
while True:
    print index,
    index += 1
    if index < 5:
        continue
    else:
        break
```

Так как цикл `while` порождает большое количество ошибок, я рекомендую использовать цикл `for`. Если программа обрабатывает множество значений и порождает новое множество значений или же сводит множество значений к одному, то нет необходимости знать номер текущего элемента, нужно лишь знать его значение. Цикл `for` итерирует по последовательности, перебирая все элементы, а значение текущего элемента присваивает переменной, объявленной в заголовке цикла.

```python
# пример порождения новой последовательности
fixed_array = []
for el in some_array:
    fixed_el = el.replace('item', 'element')
    fixed_array.append(fixed_el)
print 'Новый массив:', fixed_array
```

```python
# пример редукции последовательности к единственному значению:
factorial = 1
for num in range(10):
    factorial *= num + 1
print 'Факториал 10:', factorial
```

## Функции

Функция - это блок программы, который принимает множество элементов (аргументы) и возвращает значение.

Аргументы указываются в скобках при вызове функции.

Результат возвращается из функции с помощью оператора `return`

```python
def func(arg1, arg2):
    return arg1 + arg2
```

```python
a = 5
b = 10
func_result = func(a, b)
print 'Вызов функции func вернул', func_result
```

Аргументы, переданные в функцию, могут иметь любые названия при вызове функции, внутри функции обращение к ним будет по именам, заданным в объявлении функции (в данном случае `arg1` и `arg2`).

Внутри функции формируется локальный контекст выполнения, в котором нет доступа к глобальным переменным. 

```python
print 'Старое значение a:', a
def try_to_change_global():
    print 'Вызов функции try_to_change_global'
    a = 10 # новая переменная, значение глобальной переменной a не изменится
try_to_change_global()
print 'Новое значение a:', a
```

```python
Для доступа к глобальному контексту используется оператор global
```
